# Supervised Learning: Simple Linear Regression

```python
x = 10 * np.random.rand(100)

# print(np.random.rand(2))
```

```python
y = 3 * x + np.random.randn(100)
# print(np.random.randn(20))
```

```python
plt.scatter(x, y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression(fit_intercept=True)
```

```python
model
```

```python
X = x.reshape(-1, 1)
```

```python
model.fit(X, y)
```

```python
#weights
model.coef_
```

```python
# intercept
model.intercept_
```

```python
x_fit = np.linspace(-1, 11, 100).reshape(-1, 1)
x_fit.shape
# print(np.linspace(-1, 1, 5))
```

```python
predicted_y = model.predict(x_fit)
predicted_y.shape
```

```python
plt.scatter(x, y)
plt.plot(x_fit, predicted_y)
```

```python
from sklearn.metrics import  mean_squared_error
y_test = 3 * x + np.random.randn(100)
y_test.shape
```

```python
print("MEAN SQUARED ERROR IS: ", mean_squared_error(y_test, predicted_y))
```
